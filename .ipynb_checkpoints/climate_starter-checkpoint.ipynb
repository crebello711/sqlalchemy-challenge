```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect, and_
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
engine
```

```python
# reflect an existing database into a new model
# Declare a Base using `automap_base()`
Base = automap_base()
Base

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
measurement=Base.classes.measurement
station=Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
session
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
# Display the row's columns and data in dictionary format
first_row_measurement = session.query(measurement).first()
first_row_measurement.__dict__

first_row_station = session.query(station).first()
first_row_station
    
# Create the inspector and connect it to the engine
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for column in columns:
    print(column["name"], column["type"])
print("_________________________________________________")    
#Check the date column to validate recent date
for row in session.query(measurement.date):
    print(row.date)

#Recent Date
session.query(measurement.date).order_by(measurement.date.desc()).first()
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
last_date=session.query(measurement.date).order_by(measurement.date.desc()).first()
display(last_date)


# Calculate the date one year from the last date in data set.
query_date = dt.date(2017,8,23)-dt.timedelta(days=365)
print("Query date:",query_date)


# Perform a query to retrieve the data and precipitation scores
prcp_data=session.query(measurement.date, measurement.prcp).\
    filter(measurement.date >= query_date).all()


# Save the query results as a Pandas DataFrame and set the index to the date column
prcp_df = pd.DataFrame(prcp_data, columns=['date', 'prcp'])
#display(df.head(20))


# Sort the dataframe by date
prcp_df=prcp_df.sort_values(['date'])
prcp_df.dropna(inplace = True)
prcp_df.reset_index(drop=True,inplace=True)
#display(prcp_df)




prcp_df.set_index('date',inplace=True)
prcp_df = prcp_df.sort_index()
# Use Pandas Plotting with Matplotlib to plot the data
ax= prcp_df.plot(rot=90,figsize=(10,8))
ax.set_title('Precipitation score from "2016-08-23" to "2017-08-23"',fontsize=20)
ax.set_ylabel('Inches',fontsize=15)
ax.set_xlabel('Date',fontsize=15)
ax.legend(['precipitation'])
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
prcp_df.describe()
```

# Exploratory Station Analysis

```python
columns = inspector.get_columns('station')
for column in columns:
    print(column["name"], column["type"])
print("_________________________________________________")   

# Design a query to calculate the total number stations in the dataset
session.query(station.id).count()
```

```python

# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
sel=[measurement.station,func.count(measurement.id)]

stCount_data=session.query(*sel).\
          group_by(measurement.station).\
         order_by(func.count(measurement.id).desc()).all()
display(stCount_data)

print("Station",stCount_data[0][0],"is the most active among others")
```

```python
#for row in session.query(measurement.tobs):
    #print(row.tobs)
#Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
sel=[measurement.station,
    func.min(measurement.tobs), 
    func.max(measurement.tobs), 
    func.avg(measurement.tobs)]

activeStation = session.query(*sel).\
    filter(measurement.station==stCount_data[0][0]).\
    group_by(measurement.station).\
    order_by(measurement.station).all()
activeStation
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

tempObvt_data=session.query(measurement.tobs).\
    filter(measurement.station==stCount_data[0][0]).\
    filter(measurement.date >= query_date).all()
tempObvt_data
#converting the data to a flattened array list  
tobs = list(np.ravel(tempObvt_data))

#plot
fig1, ax1 = plt.subplots(figsize=(10, 8))

plt.xlabel("Temperature",fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.hist(tobs, bins=12)
plt.legend(["tobs"])
plt.show()
```
