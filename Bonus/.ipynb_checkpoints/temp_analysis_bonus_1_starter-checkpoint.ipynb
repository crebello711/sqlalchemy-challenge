```python
#Set the date column as the DataFrame index
df=df.set_index(df['date'])
df.head(10)
```

```python
# Drop the date column
df = df.drop(columns='date')
df
```

### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months

june_df = df.filter(like = '-06-', axis=0)
display(june_df)

december_df = df.filter(like = '-12-', axis=0)
display(december_df)
```

```python
# Identify the average temperature for June across all stations
avgTempJune = june_df.groupby('station').mean()
avgTempJune
```

```python
# Identify the average temperature for December across all stations
avgTempDecember = december_df.groupby('station').mean()
avgTempDecember
```

```python
# Create collections of temperature data
```

```python
# Run unpaired t-test
print("I will use the t-test to determine if the average temperatures between June and December are statistically significantly different. Will use the unpaired t-test because I am comparing means from separate groups by month that are not dependent on each other.")
print("______________________________________________________________________________________________________________________________")
'''
H0: The difference between the mean temperatures of June and December is equal to zero
H1: The difference between the mean temperatures of June and June and December is NOT equal to zero '''


print(stats.ttest_ind(avgTempDecember['tobs'],avgTempJune['tobs']))
print("______________________________________________________________________________________________________________________________")
print("p-value is less than 0.05, We can reject the null hypothesis and conclude that the difference in means is statistically significant in temperature between June and December.")
```
